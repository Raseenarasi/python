```python
from tkinter import*

window = Tk()

window.title("Calculator")
window.geometry('298x430')
window.configure(bg="black")

equation = ""

def show(value):
    global equation
    equation+=value
    label.config(text=equation)
    
def clear():
    global equation
    equation =""
    label.config(text=equation)
    
    
def calculate():
    global equation
    result = ""
    if equation !="":
        try:
            result=eval(equation)
        except:
            result="Error:("
            equation=""
            
        label.config(text=result)
            
            
    
label = Label(window,width=30,height=2,text="",font=("arial",30),bg="#fff",fg="black")
label.pack()

Button(window,text="C", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="blue",command=lambda:clear()).place(x=10,y=100)
Button(window,text="()",width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("()")).place(x=80,y=100)
Button(window,text="%", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("%")).place(x=150,y=100)
Button(window,text="/", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("/")).place(x=220,y=100)

Button(window,text="7", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("7")).place(x=10,y=165)
Button(window,text="8", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("8")).place(x=80,y=165)
Button(window,text="9", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("9")).place(x=150,y=165)
Button(window,text="*", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("*")).place(x=220,y=165)

Button(window,text="4", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("4")).place(x=10,y=230)
Button(window,text="5", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("5")).place(x=80,y=230)
Button(window,text="6", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("6")).place(x=150,y=230)
Button(window,text="-", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("-")).place(x=220,y=230)

Button(window,text="1", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("1")).place(x=10,y=295)
Button(window,text="2", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("2")).place(x=80,y=295)
Button(window,text="3", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("3")).place(x=150,y=295)
Button(window,text="+", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("+")).place(x=220,y=295)

Button(window,text="+/-", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("+/-")).place(x=10,y=360)
Button(window,text="0", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show("0")).place(x=80,y=360)
Button(window,text=".", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#151717",command=lambda:show(".")).place(x=150,y=360)
Button(window,text="=", width=3,height=1,font=("arial",22,"bold"),bd=2,fg="white",bg="#ED6310",command=lambda:calculate()).place(x=220,y=360)


window.mainloop()
```
